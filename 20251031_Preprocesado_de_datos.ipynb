```python
# Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importar dataset
df = pd.read_csv('datasets/Salarios_promedios.csv')
df.head()
```

```python
X = df.iloc[:,:-1].values # Nuestros datos de entrada
Y = df.iloc[:,-1].values # La prediccion etiquetada
print(X)
print(Y)
```

```python
# Libreria scikit-learn para preprocesado de datos
# Simple Imputer: Completar los datos faltantes
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(X[:, 1:3]) # Ajustar el imputer a las columnas con datos faltantes
X_imputer = X.copy()
X_imputer[:, 1:3] = imputer.transform(X[:, 1:3]) # Transformar los datos y completar los valores faltantes
print(X_imputer)
```

```python
# LabelEncoder: Convertir datos categóricos a numéricos
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X_label = X.copy()
X_label[:, 0] = labelencoder_X.fit_transform(X[:, 0]) # Convertir la primera columna (Position)
print(X_label)
```

```python
# OneHotEncoder: Convertir variable categórica a binaria
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X_transformed = X_imputer.copy()
X_transformed = np.array(ct.fit_transform(X_transformed))
print(X_transformed) 
# Notar que la primera columna se ha convertido en varias columnas binarias (en este caso 3 columnas)
```

```python
# Dividir el dataset en conjunto de entrenamiento y conjunto de testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_transformed, Y, test_size=0.2, random_state=0)
print(X_train)
```

```python
# Estandarizar datos en misma escala
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print(X_train_scaled)
```
