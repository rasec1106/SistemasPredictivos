```python
edades= [23, 45, 18, 30, 50]
# print("Todas:", edades) # Los dos puntos pueden omitirse
print("Todas:", edades[:]) # Muestra todos los elementos
print("Del 1 al ultimo:", edades[1:]) # Definimos desde el indice 1 hasta el final
print("El 1 y el 2:", edades[1:3]) # Definimos desde el indice 1 hasta el 3 (sin incluir el 3)
```

```python
# if
edad=77
if edad < 18:
    print("Menor de edad", edad)
elif edad >= 18 and edad < 65:
    print("Adulto", edad)
else:
    print("Adulto mayor", edad)
```

```python
# for
edades = [23, 45, 18, 30, 50]
for edad in edades:
    if edad < 18:
        print("Menor de edad", edad)
    elif edad >= 18 and edad < 65:
        print("Adulto", edad)
    else:
        print("Adulto mayor", edad)
```
